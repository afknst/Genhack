```python
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from Af_data_pre import save, load

sns.set_theme()
```

```python
D = 4


def read_data(_file=Path("train.csv")):
    return np.array(pd.read_csv(_file))[:, 1:]


def plot_date(_data):
    d = D

    plt.close('all')
    fig = plt.figure('Scatterplot of data points')
    for i in np.arange(d):
        for j in np.arange(d):
            ax = plt.subplot(d, d, i * (d) + j + 1)
            plt.scatter(_data[:, i], _data[:, j], 1)
            plt.tick_params(axis='both',
                            left='off',
                            top='off',
                            right='off',
                            bottom='off',
                            labelleft='off',
                            labeltop='off',
                            labelright='off',
                            labelbottom='off')
            ax.set_aspect(1)
            ax.axes.xaxis.set_visible(False)
            ax.axes.yaxis.set_visible(False)
    plt.tight_layout()
    # fig.savefig("lattice.eps", format="eps", dpi=1000)
    plt.show()

    for i in np.arange(d):
        analyze(_data[:, i], _name=str(i))

    _data_r = np.array([np.linalg.norm(x) for x in _data])
    analyze(_data_r, _name="Norm")


def analyze(_col, _name=None):
    plt.figure()
    sns.histplot(data=_col, stat='probability')

    plt.ylabel('Density')
    if _name is not None:
        plt.xlabel(_name)
        plt.title(f'Distribution of {_name}')
    plt.tight_layout()
    # plt.savefig(f'RES_{_name}_1{_format}', dpi=300)
    plt.show()

    print("Max:", np.max(_col))
    print("Min:", np.min(_col))
    print("Mean:", np.mean(_col))
    print("Std:", np.std(_col))
```
